```python
# create lists of dataframes for each metric. 
# we loop through each xml of company data for each company and extract + clean the row we want
# then, we add that dataframe slice to the list of dataframes, and use concat at the end to combine
revenue_df_list = list()

# construct revenue_df
for f in filenames:
    new_df = pd.DataFrame()
    
    income_statement = pd.read_excel(f, sheet_name='Income Statement')
    
    # retrieve the index of the row that contains "Total Revenue". 
    #     Data is not as uniform as described, so it is in different rows in some places
    total_revenue_index = income_statement.index[income_statement.iloc[:,0].str.contains('Total Revenue', na=False)][0]
    
    new_df = income_statement.iloc[[13,total_revenue_index]]
    
    # make the first row the new column name, delete old row we got names from
    new_df.columns = new_df.iloc[0]
    new_df.drop(index=new_df.index[0], axis=0, inplace=True)
    
    # drop Data Source Change column if it exists
    new_df.drop(list(new_df.filter(regex = 'Data Source Change')), axis = 1, inplace = True)

    # remove the first X characters from each item in the first row, to get just the year
    new_df.columns = [col[-4:] for col in new_df.columns]
    
    # insert the company name as a new row
    new_df.insert(1, "company", f.split('/')[1].split('.')[0])

    # rename the first column back to what it should be
    new_df.rename(columns={new_df.columns[0]: "For the Fiscal Period Ending" }, inplace=True)
    
    # remove any duplicated column names
    new_df = new_df.loc[:,~new_df.columns.duplicated()].copy()
    
    revenue_df_list.append(new_df)
    
    # display(HTML(new_df.to_html()))
    
revenue_df = pd.concat(revenue_df_list)

    
```

```python
# construct operating_income_df
operating_income_df_list = list()
for f in filenames:
    new_df = pd.DataFrame()
    
    income_statement = pd.read_excel(f, sheet_name='Income Statement')
    
    # retrieve the index of the row that contains "Operating Income". 
    #     Data is not as uniform as described, so it is in different rows in some places
    print(f)
    if ('financial_services' in f and 'Mastercard' not in f and 'Morgan_Stanley' not in f and 'Goldman' not in f and 'Visa' not in f):
        # for banks, because they are weird
        index = income_statement.index[income_statement.iloc[:,0].str.contains('Total Other Non-Interest Income', na=False)][0]
    else:
        index = income_statement.index[income_statement.iloc[:,0].str.contains('Operating Income', na=False)][0]
    
    new_df = income_statement.iloc[[13,index]]
    
    # make the first row the new column name, delete old row we got names from
    new_df.columns = new_df.iloc[0]
    new_df.drop(index=new_df.index[0], axis=0, inplace=True)
    
    # drop Data Source Change column if it exists
    new_df.drop(list(new_df.filter(regex = 'Data Source Change')), axis = 1, inplace = True)

    # remove the first X characters from each item in the first row, to get just the year
    new_df.columns = [col[-4:] for col in new_df.columns]
    
    # insert the company name as a new row
    new_df.insert(1, "company", f.split('/')[1].split('.')[0])

    # rename the first column back to what it should be
    new_df.rename(columns={new_df.columns[0]: "For the Fiscal Period Ending" }, inplace=True)
    
    # remove any duplicated column names
    new_df = new_df.loc[:,~new_df.columns.duplicated()].copy()
    
    operating_income_df_list.append(new_df)
    
    # display(HTML(new_df.to_html()))
    
operating_income_df = pd.concat(operating_income_df_list)
```

```python
# construct R&D expenses df
rd_exp_df_list = list()
for f in filenames:
    new_df = pd.DataFrame()
    
    income_statement = pd.read_excel(f, sheet_name='Income Statement')
    
    # retrieve the index of the row that contains "Operating Income". 
    #     Data is not as uniform as described, so it is in different rows in some places
    print(f)
    if ('Bank_of_America' in f or 'Citigroup' in f or 'JPM' in f):
        index = income_statement.index[income_statement.iloc[:,0].str.contains('Total Other Non-Interest Expense', na=False)][0]
    elif ('Centene' in f):
        index = income_statement.index[income_statement.iloc[:,0].str.contains('Other Operating Exp.', na=False)][0]
    elif ('Elevance' in f):
        index = income_statement.index[income_statement.iloc[:,0].str.contains('Selling and Marketing Exp.', na=False)][0]
    elif ('UnitedHealth' in f):
        index = income_statement.index[income_statement.iloc[:,0].str.contains('Selling General & Admin Exp., Total', na=False)][0]
    elif ('Morgan_Stanley' or 'Goldman' in f):
        index = income_statement.index[income_statement.iloc[:,0].str.contains('Other Non-Operating Inc.', na=False)][0]
    elif ('Home_Depot' or 'Costco' in f):
        index = income_statement.index[income_statement.iloc[:,0].str.contains('R & D Exp.', na=False)][0]
    else:
        index = income_statement.index[income_statement.iloc[:,0].str.contains('R&D Exp.', na=False)][0]   
    
    
    new_df = income_statement.iloc[[13,index]]
    
    # make the first row the new column name, delete old row we got names from
    new_df.columns = new_df.iloc[0]
    new_df.drop(index=new_df.index[0], axis=0, inplace=True)
    
    # drop Data Source Change column if it exists
    new_df.drop(list(new_df.filter(regex = 'Data Source Change')), axis = 1, inplace = True)

    # remove the first X characters from each item in the first row, to get just the year
    new_df.columns = [col[-4:] for col in new_df.columns]
    
    # insert the company name as a new row
    new_df.insert(1, "company", f.split('/')[1].split('.')[0])

    # rename the first column back to what it should be
    new_df.rename(columns={new_df.columns[0]: "For the Fiscal Period Ending" }, inplace=True)
    
    # remove any duplicated column names
    new_df = new_df.loc[:,~new_df.columns.duplicated()].copy()
    
    rd_exp_df_list.append(new_df)
    
    # display(HTML(new_df.to_html()))
    
rd_exp_df = pd.concat(rd_exp_df_list)
```
